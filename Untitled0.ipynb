```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
```

```python
from google.colab import files
uploaded=files.upload()
df=pd.read_csv('apartment_data.csv')
```

```python
df=df.dropna()
df=pd.get_dummies(df,drop_first=True)
x=df.drop('price',axis=1)
y=df['price']
```

```python
df.head()
```

```python
x=np.array(df[['buildYear']])
y=np.array(df['price'])
```

```python
plt.plot(x,y)
plt.xlabel('buildYear')
plt.ylabel('price')
plt.title('buildYear vs price')
plt.show()
```

```python
x_train=x[0:100]
y_train=y[0:100]
x_test=x[100:]
y_test=y[100:]
```

```python
scaler=StandardScaler()
x_train_scaled=scaler.fit_transform(x_train)
x_test_scaled=scaler.transform(x_test)
```

```python
model=LinearRegression()
model.fit(x_train_scaled,y_train)
```

```python
y_pred=model.predict(x_test_scaled)
mse=mean_squared_error(y_test,y_pred)
r2=r2_score(y_test,y_pred)
```

```python
print(f'mean squared error:={mse}')
print(f'R^2 Score:{r2}')
```

```python
plt.scatter(y_test,y_pred,alpha=0.5)
plt.xlabel=('true rent prices')
plt.ylabel=('predicted rent prices')
plt.title('true vs predicted rent prices')
plt.show()
```
